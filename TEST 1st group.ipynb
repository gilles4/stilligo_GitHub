```python
#drop non important columns 
selec_rslt2 = rslt_df2.drop(['WHO Region', 'ISO3', 'WHO Country Name', 
 'Reference', 'Number and type of monitoring stations', 'PM25 temporal coverage (%)', 'PM10 temporal coverage (%)','Version of the database',
       'Status'], axis=1)
selec_rslt2
```
